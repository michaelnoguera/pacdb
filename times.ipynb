```python
import polars.selectors as cs
import numpy as np

import polars as pl

import matplotlib.pyplot as plt
```

```python
fname = 'times/step1-execution_time.txt'
step1_times = [float(k[1].strip()) for k in [t.split(':') for t in open(fname, 'r').read().split('\n')][:-1]]
```

```python
fig, axes = plt.subplots(1, 1, figsize=(5, 4))
# Handle case where there's only one column (axes won't be a list)
axes = [axes]
axes[0].hist(step1_times, bins=10, alpha=0.7, edgecolor='black')
axes[0].set_title(f'Histogram of times for Step 1')
axes[0].set_xlabel('Time in seconds')
axes[0].set_ylabel('Frequency')
axes[0].grid(True, alpha=0.3)
plt.savefig(f'figs/step1_times.png', facecolor='white', bbox_inches='tight')
```
